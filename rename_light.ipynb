```python
import re
import jsonlines as jl

file = '/content/drive/MyDrive/Colab Notebooks/plantopedia_out3_clean.jl'
file_out = '/content/drive/MyDrive/Colab Notebooks/plantopedia_out4.jl'

with jl.open(file) as input:
    with jl.open(file_out, mode='w') as output:
        for item in input:
            
            new = dict()
            for el in item:
                if el == 'отношение к свету':
                    new['light'] = item[el]
                    
                    continue
                
                
                new[el.lower()] = item[el]
                
            output.write(new)
            
            
print("ended")
```
