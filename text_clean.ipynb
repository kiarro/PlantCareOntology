```python
import re
import jsonlines as jl
import unicodedata

file = '/content/drive/MyDrive/Colab Notebooks/plantopedia_out3.jl'
file_out = '/content/drive/MyDrive/Colab Notebooks/plantopedia_out3_clean.jl'

with jl.open(file) as input:
    with jl.open(file_out, mode='w') as output:
        for item in input:
            
            new = dict()
            for el in item:
                if el == 'header':
                    txt = "".join(filter(str.isalpha,item[el]))
                    new['header'] = unicodedata.normalize('NFD', txt).encode('ascii', 'ignore').decode("utf-8")
                    
                    continue
                
                
                new[el.lower()] = item[el]
            if(item['header_ru']=="Стилграсс"): new['header']="Xanthorrhoea"
                
            output.write(new)
            
            
print("ended")
```
